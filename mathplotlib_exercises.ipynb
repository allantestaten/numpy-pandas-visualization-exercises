```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = range(-100,100)
y = [(num ** 2 - num + 2) for num in x] 
```

```python
#1. Use matplotlib to plot the following equation:
#y=x**2−x+2



plt.plot(x,y)

plt.text(0,0,'Origin')
```

```python
#2. y = sqrt(x), y = x^3 , 2^x, 
 
x2 = range(100)

y2 = [math.sqrt(n) for n in x2]

plt.plot(x2,y2)

plt.text(0,0,'Origin')

```

```python
# y = x^3

x3 = np.array(range(0,20))
y3 = x3**3

plt.plot(x3,y3)

plt.text(0,0,'Origin')
```

```python
# y = 2^x

x4 = np.array(range(10))
y4 = 2**x4

plt.plot(x4,y4)

plt.text(0,0,'Origin')
```

```python
# y = 1(x+1)

x5 = np.array(range(10))

y5 = [1/(n+1) for n in x5]

plt.plot(x5,y5)

plt.text(0,0,'Origin')
```

```python
#3. Combine the figures you created in the last step into one large figure with 4 subplots.
#subplot 1
plt.figure(figsize = (10,6))

#subplot 1
plt.subplot(2,2,1)
           
plt.plot(x2,y2)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0,10)
plt.grid(True,ls = '--')          
           
#subplot 2
plt.subplot(222)

plt.plot(x3,y3)

plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--')
           
#subplot 3
plt.subplot(223)
    
plt.plot(x4,y4)
plt.grid(True, ls = '--')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.title('$y = 2^x$')

#subplot 4
plt.subplot(224)

plt.plot(x5, y5)
plt.title(r'$ y= \frac{1}{x+1}$')
plt.xlabel('$x$')
plt.ylabel(r'$ y= \frac{1}{x+1}$')
plt.grid(True, ls = '--')
plt.xlim(0,15)
plt.ylim(0)
```
